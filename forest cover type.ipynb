```python
import dask
l=[]
for feature in num_feat:
    if abs(df[[feature]].compute().skew().values[0])>=1:
        l.append(feature)
        print(feature,'=',df[[feature]].compute().skew().values[0])
```

#### as the numerical features are of different ranges------normalisation is required

```python
y=df['Cover_Type']
x=df[df.columns[:-1]]
```

```python
x.head()
```

```python
from dask_ml.preprocessing import MinMaxScaler,StandardScaler
from dask_ml.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=45)

```

```python
min_max=MinMaxScaler()
x_min_max=min_max.fit_transform(x_train)
x_min_max.head()

```

 as we can see all are in range of 0 to 1 now

#### In order to get a uniform distribution we must perform Standard scaling

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x_min_max).to_dask_array()
```
